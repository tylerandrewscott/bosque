```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
empty_district_info = pd.DataFrame()
empty_dww = pd.DataFrame()
driver = webdriver.Chrome()  # Optional argument, if not specified will search path.
base_page = 'http://www14.tceq.texas.gov/iwud/dist/index.cfm?fuseaction=ListDistricts&COMMAND=LIST&compress=N&StartName=&ID=&RegionCode=&DistTypeCode=&CreationTypeCode=&DistFunctionTypeCode=&CountyCode=&FinancialStatus=&ActivityStatus=&ListStart='
driver.get(base_page)
#CountyCode

from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
#
# select by value 
#select.select_by_value('1')
#wait = WebDriverWait(driver, 10)
select = Select(driver.find_element_by_css_selector('#CountyCode'))
wd_info_call = '.iwuddata:nth-child(2)'
#wd_info_header = 'tr:nth-child(5) .iwudheading , .iwudheading:nth-child(1)'
wd_info_header = '.iwudheading:nth-child(1)'
for county in range(1,(len(select.options)-1)):
    select_county = Select(driver.find_element_by_css_selector('#CountyCode'))
    #select_function = Select(driver.find_element_by_css_selector('#DistFunctionTypeCode'))
    select_type = Select(driver.find_element_by_css_selector('#DistTypeCode'))
    select_county.select_by_value(str(county))
    #select_function.select_by_value('STW')
    select_type.select_by_value('MUD')
    #wait = WebDriverWait(driver, 10)
    #element = wait.until(EC.element_to_be_selected((,'someid')))
    #element = wait.until(EC.element_to_be_selected((,'someid')))
    driver.find_elements_by_css_selector('input')[1].click()
    css_find_wds = "table+ table .iwud a"
    css_find_number = '.iwud+ .iwud:nth-child(3)'
    css_find_status = '.iwud~ .iwud+ .iwud'
    css_to_click_more = "table+ table tr:nth-child(1) a:nth-child(1) img"
    System = [];Status = []; ID = []
    #### Grab basic district info
    if len(css_find_wds)>0:
        for el in driver.find_elements_by_css_selector(css_find_wds):
            System.append(el.text) 
            #print(el.text)
        for el in driver.find_elements_by_css_selector(css_find_number):
            ID.append(el.text)
        for el in driver.find_elements_by_css_selector(css_find_status):
            Status.append(el.text)
        system_list_df = pd.DataFrame(
            {
            'Name': System,
             'Status': Status,
             'ID': ID,
            'County': str(county),
            'Type':'MUD'
            })
        empty_dww = empty_dww.append(system_list_df)
        water_systems = driver.find_elements_by_css_selector(css_find_wds)
        for systems in range(len(water_systems)):
            driver.find_elements_by_css_selector(css_find_wds)[systems].click()
            if any(empty_district_info.columns == driver.find_element_by_css_selector('.iwudtitle').text) == True:
                driver.back()
                continue
            info_vars = []
            info = []
            for el in driver.find_elements_by_css_selector(wd_info_call):
                info.append(el.text)
            for el in driver.find_elements_by_css_selector(wd_info_header):
                info_vars.append(el.text)
            i = 0
            for v in range(len(info_vars)):
                if info_vars[v] == '':
                    i = i + 1
                    info_vars[v] = str(i)
            #print(info_vars);print(info)
            temp_info = pd.DataFrame(info,index = info_vars)
            temp_info.columns = [driver.find_element_by_css_selector('.iwudtitle').text]
            #print(temp_info)
            empty_district_info = empty_district_info.join(temp_info,how='outer')
            grouped = empty_district_info.groupby(level=0)
            empty_district_info = grouped.last()
            driver.back()
```
