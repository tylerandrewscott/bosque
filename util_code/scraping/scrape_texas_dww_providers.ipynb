```python
import time
from selenium import webdriver
driver = webdriver.Chrome()
```

```python
harris_search = 'http://dww2.tceq.texas.gov/DWW/JSP/SearchDispatch?number=&name=&ActivityStatusCD=All&county=HARRIS&WaterSystemType=C&SourceWaterType=All&SampleType=null&begin_date=10%2F25%2F2014&end_date=10%2F25%2F2016&action=Search+For+Water+Systems'
driver.get(harris_search)
empty_df = pd.DataFrame()
summary_sheet_links = 'font a'
system_info = 'p+ table tr+ tr td'
info_names = 'p+ table tr:nth-child(1) td'
link_elements = driver.find_elements_by_css_selector(summary_sheet_links)
for link in range(len(link_elements)):
    #print(round(link/len(link_elements)))
    print(link)
    system_ids = driver.find_elements_by_css_selector('#AutoNumber7 tr+ tr td:nth-child(1)')
    print(system_ids[link].text)
    link_elements = driver.find_elements_by_css_selector(summary_sheet_links)
    link_elements[link].click()
    info = driver.find_elements_by_css_selector(system_info)
    names = driver.find_elements_by_css_selector(info_names)
    colnames = [];values = []
    for inf in info:
        #print(inf.text)
        values.append(inf.text)
    for nm in names:
        #print(nm.text)
        colnames.append(nm.text)
    temp = pd.DataFrame(values).T
    temp.columns = [colnames]
    empty_df = empty_df.append(temp)
    driver.back()
```
